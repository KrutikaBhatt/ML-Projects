```
import pandas as pd
import numpy as np

df = pd.read_csv('breast_cancer.csv')
df.head()
```

```
df =df.drop(['id'],axis=1)
df.head()
```

```
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingRegressor

df['diagnosis'] =LabelEncoder().fit_transform(df['diagnosis'])
X = df.drop('diagnosis', axis = 1)
y = df['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 92, stratify = y)
```

```
gb = GradientBoostingClassifier(random_state=1)
gb.fit(X_train,y_train)

print("Gradient Boosting Classifier ")
print("Train score :{}\nTest score :{}".format(gb.score(X_train,y_train),gb.score(X_test,y_test)))
```

```
### Running the above code in iteration to see the impact of number of trees on the output
trees = list(range(10,101))
print("Trees    Training      Test")
for tree in trees:
    gb = GradientBoostingClassifier(random_state = 1, n_estimators = tree)
    gb.fit(X_train, y_train)
    
    
    print("%d       %f      %f" % (tree, gb.score(X_train, y_train), gb.score(X_test, y_test)))
```

```
gb = GradientBoostingClassifier(
    random_state=1,
    n_estimators = 100,
    validation_fraction = 0.2,
    n_iter_no_change = 5,
    tol=0.01
)
gb.fit(X_train,y_train)
print("Gradient Boosting Results :")
print("training set score: %f" % gb.score(X_train, y_train))
print("test set score: %f" % gb.score(X_test, y_test))
print("No of trees :", gb.n_estimators_)
```

## **Grid Search**

```

from sklearn.model_selection import GridSearchCV
trees = 100
rate = [0.01,0.1,1]
depth = list(range(1,5))
split = list(range(2,4))

param_grid = { 'max_depth' : depth,
             'min_samples_split' : split,
             'learning_rate' : rate}

gb = GradientBoostingClassifier(random_state = 1, 
                                n_estimators = trees)

gb_grid_search = GridSearchCV(gb, param_grid, cv=5)

gb_grid_search.fit(X_train, y_train)
```

```
print("Grid Search Results :")
print("Train Score :{}".format(gb_grid_search.score(X_train,y_train)))
print("Test Score :{}".format(gb_grid_search.score(X_test,y_test)))
print("Best params :{}".format(gb_grid_search.best_params_))
```

```
### Let's use the best parameters
gb1 = GradientBoostingClassifier(
    random_state = 1,
    n_estimators = 100,
    max_depth = 1,
    min_samples_split = 2,
    learning_rate = 0.1,
    validation_fraction = 0.2,
    n_iter_no_change = 5, 
    tol = 0.01
)

gb1.fit(X_train, y_train)
print("Gradient Boosting Results :")
print("training set score: %f" % gb1.score(X_train, y_train))
print("test set score: %f" % gb1.score(X_test, y_test))
print("No of trees :", gb1.n_estimators_)
```
